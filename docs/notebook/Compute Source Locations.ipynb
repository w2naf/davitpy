```python
import datetime
import matplotlib.pyplot as plt
from mpl_toolkits import basemap
import pydarn
import utils
```

```python
code = 'bks'
sTime = datetime.datetime(2014,11,1)
ngates  = 60
frang   = 180.
rsep    = 45.

site    = pydarn.radar.radStruct.site(code=code,dt=sTime)
fov     = pydarn.radar.radFov.fov(frang=frang,rsep=rsep,site=site,model='IS',coords='geo',ngates=ngates)

lat_ctr =  42.8
lon_ctr = -84.2
```

```python
lst = []
lst.append({'azm':site.boresite,'r':4000,'color':'b'})
lst.append({'azm':site.boresite+90.,'r':4000,'color':'g'})

for item in lst:
    item['lat'],item['lon'] = utils.greatCircleMove(lat_ctr,lon_ctr,item['r'],item['azm'])
```
